```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("D:/ml datasets/driver-data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
df_analyze = df.drop('id', axis=1)
```

```python
kmeans.fit(df_analyze)
```

```python
kmeans.cluster_centers_
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
print(type(kmeans.labels_))
unique, counts = np.unique(kmeans.labels_, return_counts=True)
print(dict(zip(unique, counts)))
```

```python
import seaborn as sns

# Assuming df_analyze is your DataFrame

# Assign cluster labels to the DataFrame
df_analyze['cluster'] = kmeans.labels_

# Use lmplot to create a scatter plot with hue (color) based on cluster
sns.set_style('whitegrid')
sns.lmplot(x='mean_dist_day', y='mean_over_speed_perc', data=df_analyze, hue='cluster',
           palette='coolwarm', aspect=1, fit_reg=False)
```
